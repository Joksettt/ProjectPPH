**Reasoning**:
The previous code block has successfully explored the data, but the data needs to be cleaned before further analysis.  Many columns have inconsistent formatting (e.g., numbers as strings with dots and commas). I will address this now.



```python
import pandas as pd
import numpy as np

# Clean numerical columns (remove dots and commas, convert to numeric)
for col in ['persediaan_awal', 'persediaan_akhir', 'hpp_rawat_inap', 'hpp_rawat_jalan', 'total_hpp', 'pendapatan_operasional',
            'pendapatan_penjualan', 'ppn_keluaran', 'ppn_masukan', 'ppn_netto', 'laba_kotor_penjualan', 'beban_operasional',
            'laba_kotor_operasional', 'laba_sebelum_pajak', 'laba_kena_pajak', 'laba_bersih_normal', 'laba_bersih_tax_holiday']:
    df[col] = df[col].astype(str).str.replace('.', '', regex=False)
    df[col] = df[col].astype(str).str.replace(',', '.', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Display cleaned data
display(df.head())
print(df.dtypes)
print(df.isnull().sum())
```

## Data cleaning

### Subtask:
Data Cleaning: Handle Missing Values and Outliers


**Reasoning**:
Impute missing values for numerical and categorical columns, detect and handle outliers, convert data types, and perform consistency checks.



```python
import pandas as pd
import numpy as np

# Create a copy of the DataFrame to avoid modifying the original
df_cleaned = df.copy()

# 1. Missing Value Imputation
# Numerical Columns
numerical_cols = df_cleaned.select_dtypes(include=np.number).columns
for col in numerical_cols:
    df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].median())

# Categorical Columns
categorical_cols = df_cleaned.select_dtypes(exclude=np.number).columns
for col in categorical_cols:
  df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mode()[0])


# 2. Outlier Detection and Handling (using IQR method for demonstration)
for col in numerical_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned[col] = np.clip(df_cleaned[col], lower_bound, upper_bound)


# 3. Data Type Conversion (already handled in previous step)

# 4. Consistency Checks (example: check for negative values in revenue)
# Add more consistency checks as needed, based on domain knowledge
for col in ['pendapatan_penjualan', 'pendapatan_operasional', 'laba_bersih_normal']:
    if (df_cleaned[col] < 0).any():
        print(f"Warning: Negative values found in column {col}")
        df_cleaned[col] = df_cleaned[col].abs()


# Display the number of missing values before and after imputation
print("Missing values before imputation:")
print(df.isnull().sum())
print("\nMissing values after imputation:")
print(df_cleaned.isnull().sum())


# Display descriptive statistics before and after cleaning
print("\nDescriptive Statistics Before Cleaning:")
display(df.describe())

print("\nDescriptive Statistics After Cleaning:")
display(df_cleaned.describe())

display(df_cleaned.head())
```

## Data wrangling

### Subtask:
Data Wrangling: Feature Engineering and Data Transformation


**Reasoning**:
Create new features representing key financial ratios and year-over-year percentage change for key financial metrics.



```python
import pandas as pd
import numpy as np

# Create a copy to avoid modifying the original DataFrame
df_wrangled = df_cleaned.copy()

# 1. Year-over-year change calculation
for col in ['pendapatan_penjualan', 'total_hpp', 'laba_bersih_normal']:
    df_wrangled[f'YoY_{col}_Change'] = df_wrangled.groupby('tahun')[col].pct_change() * 100
    # Handle the first year (no preceding year)
    df_wrangled[f'YoY_{col}_Change'].fillna(0, inplace=True)

# 2. Ratio calculations
# Avoid division by zero
df_wrangled['Profit_Margin'] = np.where(df_wrangled['pendapatan_penjualan'] != 0,
                                        (df_wrangled['laba_bersih_normal'] / df_wrangled['pendapatan_penjualan']) * 100, 0)
df_wrangled['Operating_Margin'] = np.where(df_wrangled['pendapatan_penjualan'] != 0,
                                          (df_wrangled['pendapatan_operasional'] / df_wrangled['pendapatan_penjualan']) * 100, 0)

# Assuming 'persediaan_awal' and 'persediaan_akhir' represent current assets and liabilities (for demonstration)
df_wrangled['Current_Ratio'] = np.where(df_wrangled['persediaan_akhir'] != 0,
                                        df_wrangled['persediaan_awal'] / df_wrangled['persediaan_akhir'], 0)

# Display the first few rows of the transformed DataFrame
display(df_wrangled.head())
```
